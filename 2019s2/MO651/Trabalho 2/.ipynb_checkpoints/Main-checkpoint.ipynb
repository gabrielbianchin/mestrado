# 2. Auxiliary Functions

A translação é uma transformação linear que desloca um objeto sem alterar sua forma ou dimensão. Este deslocamento pode ser vertical, horizontal ou seguir outra orientação qualquer. A matriz de translação é representada pela forma abaixo:

$$
T_{trans} = \left[
\begin{array}{ccc}
1 & 0 & \Delta x \\
0 & 1 & \Delta y \\
0 & 0 & 1
\end{array}
\right]
$$

Uma matriz de rotação é uma transformação linear que provoca uma rotação de um objeto segundo um eixo, mantendo a suas dimensões. A matriz de rotação é representada pela forma abaixo:

$$
T_{rot} = \left[
\begin{array}{ccc}
\cos{\alpha} & -\sin{\alpha} & 0 \\
\sin{\alpha} & \cos{\alpha} & 0 \\
0 & 0 & 1
\end{array}
\right]
$$

Dessa forma, o cálculo final para determinar a $Pos_{I}$ nas coordenadas globais após uma rotação seguida de uma translação é determinado pela expressão:

$$
Pos_{I} = T_{trans} * T_{rot} * Pos_{r}
$$

```python
def localToGlobal(xObject, yObject, theta, x, y):
    xRef = xObject
    yRef = yObject
    thetaRef = theta

    tTrans = [[1, 0, xRef],
              [0, 1, yRef],
              [0, 0, 1]]

    cos_theta = cos(thetaRef)
    sin_theta = sin(thetaRef)

    tRot = [[cos_theta, -sin_theta, 0],
            [sin_theta, cos_theta, 0],
            [0, 0, 1]]

    coord = [x, y, 1]

    localToGlobalMatrix = np.dot(tTrans,tRot)
    resMatrix = np.dot(localToGlobalMatrix, coord)

    return resMatrix[0],resMatrix[1]
```

No simulador V-REP, os ângulos entre 0 até $\pi$ seguem o formato padrão. Já de $\pi$ até 0, os ângulos seguem de $-\pi$ até 0, de forma crescente.

```python
def correctEncAngle(angleEnc):
    if angleEnc >=0:
        return angleEnc
    else:
        return 2*pi+angleEnc
```

## 4.1 Atividade do corredor
Nesse desafio, o robô deve se manter ao meio de um corredor formado por duas paredes. Utilizamos o controlador PD para controlar a movimentação do robô. Testamos diversos valores para KP e KD. Os melhores resultados obtidos foram com KP igual a 1 e KD igual a 10. Avaliamos tanto o Ground Truth quanto a odometria para a realização dessa atividade.

```python
#leitura dos sensores LASER
laser = vrep.simxUnpackFloats(vrep.simxGetStringSignal(clientID, 'Laser2D', vrep.simx_opmode_streaming)[1])
#posição Ground Truth do robô
error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)
#orientação do robô
error_angle, angle = vrep.simxGetObjectOrientation(clientID, p3dx, -1, vrep.simx_opmode_streaming)

#frame global dos objetos lidos pelos sensores LASER
i = 0
while (i < int(np.shape(laser)[0])//3):
    cPX, cPY = localToGlobal(pos[0], pos[1], angle[2], laser[(i*3)+1], laser[(i*3)+2])
    laserXGT.append(cPX)
    laserYGT.append(cPY)
    i += 1
```

## 4.1.1 Ground Truth

```python
t_end = time.time() + 150
kp = 1
kd = 10

distancia = []
qtde = []
limiar = []
cont = 0

old_error = 0

error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)

while time.time() < t_end:
    error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)
    y = pos[1]
    dist = y
    meio = (laserYGT[0] + laserYGT[181]) / 2
    
    
    error = meio - dist
    diff_error = error - old_error
    old_error = error
    
    cont += 1
    
    u = error * kp + kd * diff_error
    
    distancia.append(dist)
    limiar.append(meio)
    qtde.append(pos[0])
    
    if u < 0:
        u = abs(u)
        robot.set_left_velocity(1+u)
        robot.set_right_velocity(1)
    else:
        robot.set_left_velocity(1)
        robot.set_right_velocity(1+u)
    time.sleep(1)
robot.stop()
```

## 4.2 WallFollow - Fuzzy
Nesse desafio, o robô deve se ficar a 1 metro da parede. Utilizamos o controlador Fuzzy para controlar a movimentação do robô. Avaliamos tanto o Ground Truth quanto a odometria para a realização dessa atividade.

```python
distance = ctrl.Antecedent(np.arange(0, 2, 0.1), 'distance')
velocity = ctrl.Consequent(np.arange(0, 0.1, 0.01), 'velocity')

distance.automf(number = 3, names = ['perto', 'longe menor', 'longe maior'])
distance['perto'] = fuzz.trimf(distance.universe, [0.6, 1, 1.4])
distance['longe menor'] = fuzz.trimf(distance.universe, [0, 0, 0.8])
distance['longe maior'] = fuzz.trimf(distance.universe, [1.2, 2, 2])
distance.view()

velocity.automf(number = 2, names = ['low', 'medium'])
velocity['low'] = fuzz.trimf(velocity.universe, [0, 0, 0.1])
velocity['medium'] = fuzz.trimf(velocity.universe, [0, 0.1, 0.1])
velocity.view()

#regras
rule1 = ctrl.Rule(distance['perto'], velocity['low'])
rule2 = ctrl.Rule(distance['longe menor'], velocity['medium'])
rule3 = ctrl.Rule(distance['longe maior'], velocity['medium'])

tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
```

### 4.2.1 Ground Truth

```python
t_end = time.time() + 130

distancia = []
qtde = []
limiar = []
cont = 0

flag = False

error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)

while time.time() < t_end:
    error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)
    if flag:
        y = pos[0]
        b = -3.37 #laserXGT[0] // -3.376238171187138
    else:
        y = pos[1]
        b = -4.85 #laserYGT[0] // -4.8500106113728085
    dist = abs(y-b)
    ultrassonic = robot.read_ultrassonic_sensors()
    
    aRi = robot.get_right_angle()
    aLi = robot.get_left_angle()
    
    dirr = ultrassonic[8]
    
    tipping.input['distance'] = dist
    tipping.compute()
    
    cont += 1
    
    u = tipping.output['velocity']
    
    distancia.append(dist)
    limiar.append(1)
    qtde.append(cont)
    
    value = (0.2*ultrassonic[2]) + ((0.3*ultrassonic[3]) + (0.3*ultrassonic[4])) + (0.2*ultrassonic[5])
    if value < 1.0:
        esq = ultrassonic[0]
        dirr = ultrassonic[7]
        if esq >= 3.0:
            robot.set_left_velocity(-2.0)
            robot.set_right_velocity(2.0)
            time.sleep(1)
            flag = True
        elif dirr > 3.0:
            robot.set_left_velocity(2.0) 
            robot.set_right_velocity(-2.0)
            time.sleep(1)
        else:
            robot.set_left_velocity(-1.0)
            robot.set_right_velocity(1.0)
            time.sleep(1)
    else:
        if dist > 1:
            robot.set_left_velocity(1+u)
            robot.set_right_velocity(1)
        else:
            robot.set_left_velocity(1)
            robot.set_right_velocity(1+u)
    time.sleep(1)
robot.stop()
```

## 4.3 Atividade de desviar de obstáculos com PD

Nessa atividade o robô deverá desviar-se dos obstáculos ques estão presentes na cena, onde o mesmo deve se manter a uma distância, segundo um limiar, dos objetos da cena.

### 4.3.1 Ground Truth

```python
t_end = time.time() + 150
kp = 5
kd = 20

distancia = []
qtde = []
limiar = []
cont = 0

old_error = 0

error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)

while time.time() < t_end:
    error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)
    
    ultrassonic = robot.read_ultrassonic_sensors()
    
    x = pos[0]
    y = pos[1]
    b = np.min([ultrassonic[2],ultrassonic[3],ultrassonic[4],ultrassonic[5]])
    dist = abs(x-b)
    
    error = 1 - dist
    diff_error = error - old_error
    old_error = error
    
    cont += 1
    
    u = error * kp + kd * diff_error
    
    distancia.append(dist)
    limiar.append(1)
    qtde.append(cont)
    
    
    
    value = (0.2*ultrassonic[2]) + ((0.3*ultrassonic[3]) + (0.3*ultrassonic[4])) + (0.2*ultrassonic[5])
    if value > 3.0:
        robot.set_left_velocity(1.0) #rad/s
        robot.set_right_velocity(1.0)
        time.sleep(2)
    else:
        esq = ultrassonic[0]
        dirr = ultrassonic[7]
        if esq >= 5.0 or (esq >=dirr and esq> 3.0):
            robot.set_left_velocity(-1.0) 
            robot.set_right_velocity(1.0+u)
            time.sleep(1)
        elif dirr >= 5.0 or (dirr >= esq and dirr > 3.0):
            robot.set_left_velocity(1.0+u) 
            robot.set_right_velocity(-1.0)
            time.sleep(1)
        else:
            robot.set_left_velocity(-1.0) 
            robot.set_right_velocity(1.0+u)
            time.sleep(2)
    robot.stop()
robot.stop()
```

### 4.3.2 Odometry

```python
t_end = time.time() + 150

kp = 5
kd = 20
old_error = 0

distancia = []
distreal = []
qtde = []
limiar = []
odoX = []
odoY = []
GTX = []
GTY = []
cont = 0

R = robot.WHEEL_RADIUS # 0.0975
L = robot.L # 0.36205
j = 0
Vl = Vr = 1

error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)
error_angle, angle = vrep.simxGetObjectOrientation(clientID, p3dx, -1, vrep.simx_opmode_streaming)
E = [pos[0],pos[1],angle[2]]

start_time = time.time()

while time.time() < t_end:
    error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)
    error_angle, angle = vrep.simxGetObjectOrientation(clientID, p3dx, -1, vrep.simx_opmode_streaming)

    angleEncLeft = vrep.simxGetJointPosition(clientID, leftMotorHandle, vrep.simx_opmode_streaming)[1]
    angleEncRight = vrep.simxGetJointPosition(clientID, rightMotorHandle, vrep.simx_opmode_streaming)[1]
    t = time.time() - start_time
        
    start_time = time.time()
    angleEncLeft = correctEncAngle(angleEncLeft)
    angleEncRight = correctEncAngle(angleEncRight)

    if 0 <= angleEncLeft <= pi and pi <= iniLeftAngle <= 2*pi or angleEncLeft > iniLeftAngle:
        leftDiff = pi - abs(abs(angleEncLeft - iniLeftAngle) - pi)
    else:
        leftDiff = -(pi - abs(abs(angleEncLeft - iniLeftAngle) - pi))

    if 0 <= angleEncRight <= pi and pi <= iniRightAngle <= 2*pi or angleEncRight > iniRightAngle:
        rightDiff = pi - abs(abs(angleEncRight - iniRightAngle) - pi)
    else:
        rightDiff = -(pi - abs(abs(angleEncRight - iniRightAngle) - pi))

    Vl = R*leftDiff/t
    Vr = R*rightDiff/t

    S = ((Vr+Vl)/2)*t
    angT = ((Vr-Vl)/(L))*t

    iniLeftAngle = angleEncLeft
    iniRightAngle = angleEncRight

    E = np.add(E, [S*cos(E[2]+angT/2),S*sin(E[2]+angT/2),angT])
    
    odoX.append(E[0])
    odoY.append(E[1])
    GTX.append(pos[0])
    GTY.append(pos[1])
    
    ultrassonic = robot.read_ultrassonic_sensors()
    
    x = E[0]
    y = E[1]
    b = np.min([ultrassonic[2],ultrassonic[3],ultrassonic[4],ultrassonic[5]])
    dist = abs(x-b)
    
    error = 1 - dist
    diff_error = error - old_error
    old_error = error
    
    cont += 1
    
    u = error * kp + kd * diff_error
    
    distreal.append(abs(pos[1] - b))
    distancia.append(dist)
    limiar.append(1)
    qtde.append(cont)
    
    value = (0.2*ultrassonic[2]) + ((0.3*ultrassonic[3]) + (0.3*ultrassonic[4])) + (0.2*ultrassonic[5])
    if value > 3.0:
        robot.set_left_velocity(1.0) #rad/s
        robot.set_right_velocity(1.0)
        time.sleep(2)
    else:
        esq = ultrassonic[0]
        dirr = ultrassonic[7]
        if esq >= 5.0 or (esq >=dirr and esq> 3.0):
            robot.set_left_velocity(-1.0) 
            robot.set_right_velocity(1.0+u)
            time.sleep(1)
        elif dirr >= 5.0 or (dirr >= esq and dirr > 3.0):
            robot.set_left_velocity(1.0+u) 
            robot.set_right_velocity(-1.0)
            time.sleep(1)
        else:
            robot.set_left_velocity(-1.0) 
            robot.set_right_velocity(1.0+u)
            time.sleep(2)
    robot.stop()
robot.stop()
```

```python
plt.plot(qtde, limiar, color = 'red', label = 'Limiar')
plt.plot(qtde, distancia, color = 'blue', label = 'Distância medida')
plt.plot(qtde, distreal, color = 'green', label = 'Distância real')
plt.xlabel('Tempo')
plt.ylabel('Distância')
plt.legend(loc = 'best')
plt.show()
```

O gráfico mostra os valores calculados pela odometria(em azul) do robô para este experimento. Neste exemplo, nota-se que os valores da odometria não foram tão divergentes dos valores reais. Por mais que o ambiente exigesse uma grande quantidades de 'curvas' pelo robô.

## 4.4 Atividade de desviar de obstáculos com FUZZY

Nessa atividade o robô deverá desviar-se dos obstáculos ques estão presentes na cena, onde o mesmo deve se manter a uma distância, segundo um limiar, dos objetos da cena.

```python
def createSensorAtecendent(name: str):
    sensor = ctrl.Antecedent(np.arange(0, 2.01, 0.01), name)
    sensor['very close'] = fuzz.trapmf(sensor.universe, [0, 0, 0.2, 0.3])
    sensor['close'] = fuzz.trimf(sensor.universe, [0.2, 0.3, 0.5])
    sensor['within_range'] = fuzz.trimf(sensor.universe, [0.3, 0.6, 0.9])
    sensor['far'] = fuzz.trapmf(sensor.universe, [0.8, 0.9, 2, 2])
    return sensor
```

```python
def createLinearSpeedConsequent():
    speed = ctrl.Consequent(np.arange(-0.1, 0.3, 0.01), 'linearSpeed')
    speed['back'] = fuzz.trimf(speed.universe, [-0.1, -0.03, 0])
    speed['stop'] = fuzz.trimf(speed.universe, [-0.01, 0, 0.01])
    speed['slow'] = fuzz.trimf(speed.universe, [0.03, 0.09, 0.15])
    speed['fast'] = fuzz.trapmf(speed.universe, [0.1, 0.2, 0.3, 0.3])
    speed.defuzzify_method = 'centroid'
    return speed
```

```python
def createAngularSpeedConsequent():
    angularSpeed = ctrl.Consequent(np.arange(-0.5, 0.51, 0.01), 'angularSpeed')
    angularSpeed.automf(7, "quant", ["verySharpRight", "sharpRight", "right", "straight", "left", "sharpLeft", "verySharpLeft"])
    angularSpeed.defuzzify_method = 'mom'
    return angularSpeed
```

```python
def createRules(frontSensors: List[ctrl.Antecedent],
                leftSensors: List[ctrl.Antecedent],
                rightSensors: List[ctrl.Antecedent],
                linearSpeed: ctrl.Consequent,
                angularSpeed: ctrl.Consequent):
    rules = [

        ctrl.Rule(frontSensors[0]['far'] & frontSensors[1]['far'], angularSpeed['straight']),
        ctrl.Rule(frontSensors[0]['far'] & frontSensors[1]['far'], linearSpeed['fast']),

        ctrl.Rule(frontSensors[0]['within_range'] | frontSensors[1]['within_range'], linearSpeed['slow']),
        ctrl.Rule(frontSensors[0]['within_range'] | frontSensors[1]['within_range'], angularSpeed['straight']),

        ctrl.Rule(frontSensors[0]['close'] | frontSensors[1]['close'], linearSpeed['stop']),
        ctrl.Rule(frontSensors[0]['close'] | frontSensors[1]['close'], angularSpeed['left']),

        ctrl.Rule(frontSensors[0]['very close'] | frontSensors[1]['very close'], linearSpeed['back']),
        ctrl.Rule(frontSensors[0]['very close'] | frontSensors[1]['very close'], angularSpeed['sharpLeft']),

    ]
    for sensor in leftSensors:
        rules.append(ctrl.Rule(sensor['close'], angularSpeed['sharpRight']))
        rules.append(ctrl.Rule(sensor['very close'], angularSpeed['verySharpRight']))
        rules.append(ctrl.Rule(sensor['far'], angularSpeed['left']))

    for sensor in rightSensors:
        rules.append(ctrl.Rule(sensor['close'], angularSpeed['sharpLeft']))
        rules.append(ctrl.Rule(sensor['very close'], angularSpeed['verySharpLeft']))
        rules.append(ctrl.Rule(sensor['far'], angularSpeed['right']))

    return rules
```

```python
class FuzzyAvoidObstacle():

    def __init__(self):
        sensors = []
        for i in range(0, 8):
            sensors.append(createSensorAtecendent('sensor' + str(i)))

        self.linearCsq = createLinearSpeedConsequent()
        self.angularCsq = createAngularSpeedConsequent()

        self.rules = createRules([sensors[3], sensors[4]],
                                 [sensors[0], sensors[1], sensors[2]],
                                 [sensors[5], sensors[6], sensors[7]],
                                 self.linearCsq, self.angularCsq)

        self.ctrl = ctrl.ControlSystem(self.rules)
        self.sys = ctrl.ControlSystemSimulation(self.ctrl)
        self.sensors = sensors

    def compute(self, sensorsReadings: List[float]):
        logging.debug(sensorsReadings)
        for i in range(0, 8):
            self.sys.input['sensor' + str(i)] = sensorsReadings[i]

        self.sys.compute()
        return self.sys.output['linearSpeed'], self.sys.output['angularSpeed']
```

```python
def vRToDrive(vLinear, vAngular):
        return (((2 * vLinear) + (robot.ROBOT_WIDTH * vAngular)) / (2*robot.WHEEL_RADIUS));

def vLToDrive(vLinear, vAngular):
    return (((2 * vLinear) - (robot.ROBOT_WIDTH * vAngular)) / (2*robot.WHEEL_RADIUS));
```

### 4.4.1 Ground Truth

```python
t_end = time.time() + 150

distancia = []
qtde = []
limiar = []
cont = 0

error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)

controller = FuzzyAvoidObstacle()

while time.time() < t_end:
    error_pos, pos = vrep.simxGetObjectPosition(clientID, p3dx, -1, vrep.simx_opmode_streaming)
    
    ultrassonic = robot.read_ultrassonic_sensors()
    
    x = pos[0]
    y = pos[1]
    b = np.min([ultrassonic[2],ultrassonic[3],ultrassonic[4],ultrassonic[5]])
    dist = abs(x-b)
    
    cont += 1
    
    distancia.append(dist)
    limiar.append(1)
    qtde.append(cont)
    
    lspeed, aspeed = controller.compute(ultrassonic)
    
    robot.set_left_velocity(vLToDrive(lspeed, aspeed))
    robot.set_right_velocity(vRToDrive(lspeed, aspeed))
    time.sleep(2.0)
    
robot.stop()
```

```python
plt.plot(qtde, limiar, color = 'red', label = 'Limiar')
plt.plot(qtde, distancia, color = 'blue', label = 'Distância medida')
plt.xlabel('Tempo')
plt.ylabel('Distância')
plt.legend(loc = 'best')
plt.show()
```

A figura mostra as posições do robô em relação à distância dos objetos presentes na cena, onde essa distância é delimitada pelo limiar que neste caso é igual a um (1). É possível notar uma melhora em relação ao experimento 4.3.1, onde o robô consegue-se manter mais, e por mais tempo, distante e dos obstáculos.
